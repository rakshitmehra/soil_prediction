```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import pickle
import joblib
```

```python
y = df['Label']
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify = y)
```

## Some new model

```python
# Stacked
base_models = [
    ('rf', RandomForestClassifier(n_estimators=50, random_state=42)),  # Random Forest
    ('gb', GradientBoostingClassifier(n_estimators=50, random_state=42)),  # Gradient Boosting
    ('svm', SVC(probability=True, random_state=42))  # Support Vector Machine
]

# Define Stacking Classifier with Logistic Regression as meta-learner
stacked_model = StackingClassifier(estimators=base_models, final_estimator=LogisticRegression())

# Train the stacked model
stacked_model.fit(X_train, y_train)

# Evaluate the model
y_pred = stacked_model.predict(X_test)
stacked_accuracy = accuracy_score(y_test, y_pred)

# Return accuracy
stacked_accuracy
```

```python
# Voting
rf = RandomForestClassifier(n_estimators=100, random_state=42)
xgb = XGBClassifier(eval_metric='logloss', random_state=42)
svm = SVC(probability=True, kernel='rbf', random_state=42)

# Assign weights based on cross-validation accuracy
weights = []
for model in [rf, xgb, svm]:
    scores = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy')
    weights.append(scores.mean())

# Normalize weights
weights = np.array(weights) / np.sum(weights)

# Define Weighted Voting Classifier
ensemble = VotingClassifier(estimators=[
    ('RandomForest', rf),
    ('XGBoost', xgb),
    ('SVM', svm)
], voting='soft', weights=weights)

# Train ensemble model
ensemble.fit(X_train, y_train)

# Predictions
y_pred = ensemble.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f'Ensemble Model Accuracy: {accuracy * 100:.2f}%')
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Hybrid Stacking Ensemble Model

# Hybrid Stacking Ensemble Model that combines Random Forest, XGBoost, and LightGBM as base learners with Logistic Regression as the meta-learner. This model dynamically learns which base model performs best for different parts of the dataset. This hybrid stacking ensemble model improves prediction accuracy by leveraging multiple models in a meta-learning approach. It dynamically learns which base model to trust more for different parts of the dataset. This approach adds novelty compared to just using a single model like Random Forest.

from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_val_predict

# Define base learners
rf = RandomForestClassifier(n_estimators=100, random_state=42)
xgb = XGBClassifier(n_estimators=100, random_state=42, eval_metric='mlogloss')
lgb = LGBMClassifier(n_estimators=100, random_state=42)

# Train base learners
rf.fit(X_train, y_train)
xgb.fit(X_train, y_train)
lgb.fit(X_train, y_train)

# Generate predictions for stacking
rf_preds = cross_val_predict(rf, X_train, y_train, cv=3, method='predict_proba')
xgb_preds = cross_val_predict(xgb, X_train, y_train, cv=3, method='predict_proba')
lgb_preds = cross_val_predict(lgb, X_train, y_train, cv=3, method='predict_proba')

# Stack predictions as new features
stacked_train = np.hstack((rf_preds, xgb_preds, lgb_preds))
rf_test_preds = rf.predict_proba(X_test)
xgb_test_preds = xgb.predict_proba(X_test)
lgb_test_preds = lgb.predict_proba(X_test)
stacked_test = np.hstack((rf_test_preds, xgb_test_preds, lgb_test_preds))

# Train meta-learner
meta_learner = LogisticRegression()
meta_learner.fit(stacked_train, y_train)

# Predict using the meta-learner
y_pred = meta_learner.predict(stacked_test)

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
print(f'Stacking Ensemble Accuracy: {accuracy:.4f}')
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Define Base Models with blending

# This hybrid ensemble model combines stacking and blending techniques to improve soil type prediction accuracy. It leverages multiple base models and a meta-learner to dynamically determine the best model for different dataset regions.

#  Implemented a hybrid ensemble approach combining stacking and blending, using Random Forest, XGBoost, and SVM with a logistic regression meta-learner.

base_models = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('xgb', XGBClassifier(eval_metric='mlogloss', random_state=42)),
    ('svm', SVC(probability=True, random_state=42))
]

# Stacking Classifier with Meta-Learner
stacking_model = StackingClassifier(estimators=base_models, final_estimator=LogisticRegression())
stacking_model.fit(X_train, y_train)

# Blending Approach
rf_pred = base_models[0][1].fit(X_train, y_train).predict_proba(X_test)
xgb_pred = base_models[1][1].fit(X_train, y_train).predict_proba(X_test)
svm_pred = base_models[2][1].fit(X_train, y_train).predict_proba(X_test)

# Weighted Average for Blending
blended_pred = (0.4 * rf_pred + 0.4 * xgb_pred + 0.2 * svm_pred).argmax(axis=1)

# Evaluate Model
stacking_acc = accuracy_score(y_test, stacking_model.predict(X_test))
blending_acc = accuracy_score(y_test, blended_pred)

print(f'Stacking Model Accuracy: {stacking_acc:.4f}')
print(f'Blending Model Accuracy: {blending_acc:.4f}')
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Evaluate Model
stacking_acc = accuracy_score(y_test, stacking_model.predict(X_test))
blending_acc = accuracy_score(y_test, blended_pred)

# Detailed metrics for blending approach
print("MULTI-METHOD ENSEMBLE MODEL CLASSIFICATION REPORT:")
print(classification_report(y_test, blended_pred))

# Confusion matrix for blending approach
print("Confusion Matrix:")
print(confusion_matrix(y_test, blended_pred))
```
